```python
import matplotlib.pyplot as plt
x=df['source_id']
y=df['market_cap']
plt.xlabel('source_id')
plt.ylabel('market_cap')
plt.plot(x,y)
plt.show()
```

```python
x=df['source_id']
y=df['market_cap']
plt.scatter(x,y,label="market_cap")
plt.xlabel("source_id")
plt.ylabel("market_cap")
plt.legend()
plt.show()
```

```python
x=df['name']
y=df['market_cap']
plt.scatter(x,y,label="market_cap")
plt.xlabel("name")
plt.ylabel("market_cap")
plt.legend()
plt.show()
```
